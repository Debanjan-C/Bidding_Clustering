### Data Dictionary

- Record ID: (int data type), not null, unique, it contains a unique set of record ids for each bidding record in the dataset. 
- Auction ID: (int data type), not null, unique, it contains a unique set of auction ids for each  record in the dataset. 
- Bidder ID: (int data type), not null, unique, it contains a unique set of ids for each bidder in the dataset.
- Bidder Tendency: (float data type), not null, it checks a score of how many bids a bidder would participate in. If someone is shill bidding then they would participate in a few sales witht the same sellers in usual cases. 
- Bidding Ratio: (float data type), not null, this will check how frequently or the ratio of how often an individual participates in a bid. Those who participate more frequently could be shill bidders who try to elevate the auction price for other bidders who may be making authentic bids.
- Successive Outbidding: (float data type), not null, This checks a score on how much a bidder may try to outbid himself even though he may be the current winner and would want to make a small and steady increase on price. 
- Last Bidding: (float data type), not null, This will check how much an individual would be active on their last bid as a shill bidder could sometimes be inactive towardss the end when more thn 90% of the auction has gone by to lose a bid and it could be part of their strategy. 
- Auction Bids: (float data type), not null, This shows us a number of auction bids that an user may have done and a shill bidder would usually have a higher number of bids than the average bids in auctions that may be going on at the same time.
- Starting_Price_Average:(float data type), not null, This will be the starting price that each of the bidders may propose and a shill bidder may usually ask for a smaller starting price to attract all authentic biddders to come and make claims on bidding.
- Early Bidding: (float data type), not null, This will show at what percentage ofthe overall bidding time the bid was called and ashill bidders usually bid early usually within the first 25% of the time.
- Winning Ratio: (float data type), not null, This shows us the ratio of the proprion of bids that users won. Some shill bidders compete in many auctions and seem to win in barely any of them. 
- Auction Duration: (int data type), not null, The duration of how long an auction lasted for.
- Class: (int data type), not null, this is the target data type where we eill see if a bidding is considered authentic or not. Marked with 0 or 1. However, we will be dropping this column for the project as we do not use target variables in unsupervised learning algortihms.

### Feature Engineering

**In this step, we are converting categorical to to dummy values using label encoding.**

### PCA - Principal Component Analysis

Below we are using the concept of Principal Componenet Analysis or PCA to reduce the dimensions of the dataset and that could bring more clarity to the dataset for the steps later on. PCA would also help reduce redundancy in the dataset. We have a large dataset and it seems there are many factors that could affect clustering like the winning ratios and all, so I felt PCA would help in reducing the dimensions and redundancy for a large dataset. I chose a score of 98 as a much lower score than that could affect your dataset as you may lose a lot of data in the dimensionality reduction and reduce any redundancy.

```python
from sklearn.decomposition import PCA
pca = PCA(n_components = 0.98)
df_pca = pca.fit_transform(scaler_df)
```

Below is a curve that shows us the cumulative variance ratio for the dataset in PCA. The x axis shows us the number of components in the dataset and the y axis contains the cumulative explianed variance. 
The curve showws us that up from 0 - 1 compnents we are able to retain 40% of the variance and we need more than 10 (12) componenets to reach a close within the 90 - 100% varaince range. We also have the sum of all of the explained variance ratio that we see on the graph above.

```python
pca = PCA().fit(scaler_df.data)
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('number of components')
plt.ylabel('cumulative explained variance');
```

```python
pca.explained_variance_ratio_.sum()
```

### KMeans Clustering

For Clustering, we will be using KMeans clustering algorithms where we group our datas into different categories in order to evaluate or get a specific result that may indicate something. WE will use the features to predict our potential results as this is unsupervised learning and we are not provided with the data. In order to get an idea of the results, I used K-Means algortihm as this will find out the center points and evaluate if the cluster distances are closer to that and I feel it is an **advantage**. The **disadvantage** is that it may not be as informative like hierarchichal clustering where they take the realtionship of each individual feautre into consideration before proceeding further.

```python
from sklearn.cluster import KMeans
```

In the first step, we will be instantiating the algorithmss, then fitting the model into our dataset and then making a prediction for each of the points. We will also assign a value for the cluster centers if we need it later on. I developed the Elbow visualizer below and 3 was the ideal score as it showed the highest sillettehouse score. In our scenario, we mainly try to figure out if a bidding was authentic or not, but there are many other factors that may not make it as obvious therefore, I feel 3 clusters would help in this case.

```python
k_means = KMeans(n_clusters= 3)
k_means.fit(df_pca)
pred = k_means.predict(df_pca)
cl_centers = k_means.cluster_centers_
```

```python
plt.scatter(df_pca[:, 0], df_pca[:, 2], c = pred, s = 10);
```

Below, we develope a Kelbow visualizaer where we use the silhouette metric to evalaute how well the cluster is performing and a higher score indicates a better defined cluster. The silhouette metric will be calculated by computing the difference between the mean of one cluster point with the other divided by the maximum of both mean scores. This helps us get an better idea on the overall clusters.

```python
##  install yellowbrck library -- pip install yellowbrick
from yellowbrick.cluster import KElbowVisualizer

# Instantiate the clustering model and visualizer
model = k_means

visualizer = KElbowVisualizer(model, k=(2,12), metric = 'silhouette', timings=False)

visualizer.fit(df_pca)        # Fit the data to the visualizer
visualizer.show()        # Finalize and render the figure
```

These are our results of the predicted values from the clusters of KMeans

```python
np.unique(pred)
```

### Agglomerative Hierarchical Clustering

In the first step, we will be instantiating the algorithmss, then fitting the model into our dataset and then making a prediction for each of the points. We will also assign a value for the cluster centers if we need it later on. In this algorithm, we are using herarchical clustering - Agglomerative. For this clustering, we look at our observations and group the most of the observations from a bottom up approrach. In comparison to Kmeans, one offf the **advantages** for this algorithm is that it takes individual relationships of the features or the data into consideration, while Kmeans looks at the overall. One of the **disadvantages** is that it can be senssitive to outliers.

```python
from sklearn.cluster import AgglomerativeClustering
h_cluster = AgglomerativeClustering(n_clusters=3)
pred_h_cluster = h_cluster.fit_predict(df_pca)
```

```python
plt.scatter(df_pca[:, 0], df_pca[:, 2], c = pred_h_cluster, s = 10);
```

These are our results of the predicted values from the clusters of Agglomerative Hierarchical Clustering.

```python
np.unique(pred_h_cluster)
```
